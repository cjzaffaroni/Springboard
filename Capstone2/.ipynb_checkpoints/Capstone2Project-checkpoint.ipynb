```python
#import os, pandas, re, matlab

import os
import pandas as pd
import re
from matplotlib import pyplot as plt
from matplotlib import ticker
```

```python
# Read file in with data for the 5 countries with the 5 variables for the last 20 years. Data pulled from the databank.worldbank.org

df = pd.read_csv('c8e65728-9429-4d9c-899b-ce34e42f8de5_Data.csv')
```

```python
# Take a look at the first few rows

print(df.head())
```

```python
#Call the info method on df to see a summary of the data

df.info()
df.describe()
```

```python
# Melt columns of date into 1 column

df = pd.melt(df, id_vars= ['Series Name', 'Series Code', 'Country Name', 'Country Code'])
```

```python
print(df.head())
print(df.tail())
```

```python
# Rename columns

df = df.rename(columns = {'variable': 'Year', 'value': 'Value'})
```

```python
#Call the info method on df to see a summary of the data after melting the data

df.info()
df.describe()
```

```python
df.dtypes
```

```python
# Find how many have null values

df.isna().sum()
```

```python
# Will drop the rows with null Series Name and Series Code

df = df.dropna(subset = ['Series Name', 'Series Code'])
```

```python
# Find how many have null values

df.isna().sum()
```

```python
# Call the info method on df to see a summary of the data after dropping nulls. I have 24 years of 5 variables 
# for 5 countries. That should be 600 rows

df.info()
df.describe()
```

```python
# Need to change Year to a date format but first take out the unnecessary data in the column

df['Year'] = df['Year'].str.replace(r'\D','',regex=True)
df['Year'] = df['Year'].str[:-4]
```

```python
# Check to see if the dates are correct with max and min

print(df['Year'])
print(df['Year'].max())
print(df['Year'].min())
```

```python
# Change format to date 

df['Year'] = pd.to_datetime(df['Year'])
```

```python
# Check the change

df.dtypes
```

```python
# Take a look at unique values of each column

print(df['Series Name'].unique())
print(df['Series Code'].unique())
print(df['Country Name'].unique())
print(df['Country Code'].unique())
print(df['Year'].unique())

```

```python
# Check the shape of the data set after changes

df.shape
```

```python
# Collecting Data for chart of GDP per country

series =  ['GDP (constant 2015 US$)']
df_GDP = df[df['Series Name'].isin(series)]
df_GDP_2 = df_GDP.sort_values(['Country Code','Year'],ascending = [True,True])
df_GDP_2['Value_pct_change'] = (df_GDP_2['Value'].pct_change()) * 100
year =  ['1998-01-01']
df_GDP_3 = df_GDP_2[~df_GDP_2['Year'].isin(year)]
df_GDP_3

df_GDP_3.groupby(['Year','Country Code'])['Value_pct_change'].sum().unstack().plot.bar()
plt.ylabel('GDP constant USD (Y/Y) %')
plt.show();
```

```python
# Collecting Data for chart of Consumption per country

series =  ['Final consumption expenditure (constant 2015 US$)']
df_Con = df[df['Series Name'].isin(series)]
df_Con_2 = df_Con.sort_values(['Country Code','Year'],ascending = [True,True])
df_Con_2['Value_pct_change'] = (df_Con_2['Value'].pct_change()) * 100
year =  ['1998-01-01']
df_Con_3 = df_Con_2[~df_Con_2['Year'].isin(year)]
df_Con_3

df_Con_3.groupby(['Year','Country Code'])['Value_pct_change'].sum().unstack().plot.bar()
plt.ylabel('Consumption constant USD (Y/Y) %')
plt.show();
```

```python
# Collecting Data for chart of FDI per country

series =  ['Foreign direct investment, net inflows (BoP, current US$)']
df_FDI = df[df['Series Name'].isin(series)]
df_FDI_2 = df_FDI.sort_values(['Country Code','Year'],ascending = [True,True])
df_FDI_2['Value_pct_change'] = (df_FDI_2['Value'].pct_change()) * 100
year =  ['1998-01-01']
df_FDI_3 = df_FDI_2[~df_FDI_2['Year'].isin(year)]
df_FDI_3

df_FDI_3.groupby(['Year','Country Code'])['Value_pct_change'].sum().unstack().plot.bar()
plt.ylabel('FDI USD (Y/Y) %')
plt.show();
```

```python
# Collecting Data for chart of Population per country

series =  ['Population, total']
df_Pop = df[df['Series Name'].isin(series)]
df_Pop_2 = df_Pop.sort_values(['Country Code','Year'],ascending = [True,True])
df_Pop_2['Value_pct_change'] = (df_Pop_2['Value'].pct_change()) * 100
year =  ['1998-01-01']
df_Pop_3 = df_Pop_2[~df_Pop_2['Year'].isin(year)]
df_Pop_3

df_Pop_3.groupby(['Year','Country Code'])['Value_pct_change'].sum().unstack().plot.bar()
plt.ylabel('Population level Y/Y')
plt.show();
```

```python
# Collecting Data for chart of Life Expectancy per country

series =  ['Life expectancy at birth, total (years)']
df_Lif = df[df['Series Name'].isin(series)]
df_Lif_2 = df_Lif.sort_values(['Country Code','Year'],ascending = [True,True])
df_Lif_2['Value_pct_change'] = (df_Lif_2['Value'].pct_change()) * 100
year =  ['1998-01-01']
df_Lif_3 = df_Lif_2[~df_Lif_2['Year'].isin(year)]
df_Lif_3

df_Lif_3.groupby(['Year','Country Code'])['Value_pct_change'].sum().unstack().plot.bar()
plt.ylabel('Life Expectancy Y/Y')
plt.show();
```

```python
# Create a data series for each Country to forecast life expectancy with 
# GDP (Y/Y), Consumption (Y/Y), FDI (Y/Y), Population

# Consider dropping the last 2 years: 2020 and 2021 because of the effect the COVID pandemia had on these numbers


```

```python
# Start with ARG

series =  ['ARG']
df_ARG = df[df['Country Code'].isin(series)]
df_ARG_2 = df_ARG.sort_values(['Series Name','Year'],ascending = [True,True])


#Use the loc accessor on df_ARG to print the 'Value' columns for GDP only
df_ARG_3 = df_ARG_2.loc[df_ARG_2['Series Name'] == 'GDP (constant 2015 US$)', ['Year', 'Value']]

df_ARG_4 = df_ARG_3.rename(columns = {'Value': 'GDP'})
df_ARG_4

# Calculate percent change

df_ARG_4['GDP_Y_Y'] = (df_ARG_4['GDP'].pct_change()) * 100
year =  ['1998-01-01']
df_ARG_GDP = df_ARG_4[~df_ARG_4['Year'].isin(year)]
df_ARG_GDP
```

```python
#Use the loc accessor on df_ARG to print the 'Value' columns for Consumption only
df_ARG_3 = df_ARG_2.loc[df_ARG_2['Series Name'] == 'Final consumption expenditure (constant 2015 US$)', ['Year', 'Value']]

df_ARG_4 = df_ARG_3.rename(columns = {'Value': 'CON'})
df_ARG_4

# Calculate percent change

df_ARG_4['CON_Y_Y'] = (df_ARG_4['CON'].pct_change()) * 100
year =  ['1998-01-01']
df_ARG_CON = df_ARG_4[~df_ARG_4['Year'].isin(year)]
df_ARG_CON
```

```python
#Use the loc accessor on df_ARG to print the 'Value' columns for FDI only
df_ARG_3 = df_ARG_2.loc[df_ARG_2['Series Name'] == 'Foreign direct investment, net inflows (BoP, current US$)', ['Year', 'Value']]

df_ARG_4 = df_ARG_3.rename(columns = {'Value': 'FDI'})
df_ARG_4

# Calculate percent change

df_ARG_4['FDI_Y_Y'] = (df_ARG_4['FDI'].pct_change()) * 100
year =  ['1998-01-01']
df_ARG_FDI = df_ARG_4[~df_ARG_4['Year'].isin(year)]
df_ARG_FDI
```

```python
#Use the loc accessor on df_ARG to print the 'Value' columns for Population only
df_ARG_3 = df_ARG_2.loc[df_ARG_2['Series Name'] == 'Population, total', ['Year', 'Value']]

df_ARG_4 = df_ARG_3.rename(columns = {'Value': 'POP'})
df_ARG_4

# Calculate percent change

df_ARG_4['POP_Y_Y'] = (df_ARG_4['POP'].pct_change()) * 100
year =  ['1998-01-01']
df_ARG_POP = df_ARG_4[~df_ARG_4['Year'].isin(year)]
df_ARG_POP
```

```python
#Use the loc accessor on df_ARG to print the 'Value' columns for Life Expectancy only
df_ARG_3 = df_ARG_2.loc[df_ARG_2['Series Name'] == 'Life expectancy at birth, total (years)', ['Year', 'Value']]

df_ARG_4 = df_ARG_3.rename(columns = {'Value': 'EXP'})
df_ARG_4

# Calculate percent change

df_ARG_4['EXP_Y_Y'] = (df_ARG_4['EXP'].pct_change()) * 100
year =  ['1998-01-01']
df_ARG_EXP = df_ARG_4[~df_ARG_4['Year'].isin(year)]
df_ARG_EXP
```

```python
# Join all dataframes into one dataframe

ARG = pd.merge(df_ARG_GDP, df_ARG_CON, on='Year')
ARG = pd.merge(ARG, df_ARG_FDI, on='Year')
ARG = pd.merge(ARG, df_ARG_POP, on='Year')
ARG = pd.merge(ARG, df_ARG_EXP, on='Year')
ARG
```
